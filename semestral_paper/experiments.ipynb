```python
# IMporty

import pandas as pd
from sim_cafe import CafeSimulation
```

## Monte Carlo - jeden sc√©n√°≈ô

```python
# MonteCarlo - jeden scenar

CONFIG = "sim_config_barista_2_2.yaml"
CYCLES = 10

print("\n" + "=" * 60)
print(f"P≈ò√çKLAD {CYCLES} opakov√°n√≠ simulace (Monte Carlo)")
print("=" * 60)

results = []

for run_id in range(CYCLES):
    print(f"\nBƒõh {run_id + 1}/{CYCLES}...")

    sim = CafeSimulation(CONFIG)
    sim.run()

    metrics = sim.get_last_metrics()
    metrics["run_id"] = run_id
    results.append(metrics)

    # P≈ôev√©st na DataFrame
    df_results = pd.DataFrame(results)

print("\n" + "=" * 60)
print(f"SHRNUT√ç {CYCLES} BƒöH≈Æ:")
print("=" * 60)

print("\nPr≈Ømƒõrn√© hodnoty:")
print(df_results[["total_served", "total_reneged", "success_rate"]].mean())

print("\nSmƒõrodatn√© odchylky:")
print(df_results[["total_served", "total_reneged", "success_rate"]].std())
```

## Monte Carlo - v√≠ce sc√©n√°≈ô≈Ø

```python
# Monte Carlo - vice scenaru

print("\n" + "=" * 60)
print("Porovn√°n√≠ r≈Øzn√Ωch konfigurac√≠")
print("=" * 60)

scenarios = {
    "Barista smƒõny: 1>1 2>2": "sim_config_barista_1_2.yaml",
    "Barista smƒõny: 1>2 2>1": "sim_config_barista_2_1.yaml",
    "Barista smƒõny: 1>2 2>2": "sim_config_barista_2_2.yaml",
}

scenario_results = []

for scenario_name, config_file in scenarios.items():
    print(f"\nSc√©n√°≈ô: {scenario_name}")

    try:
        sim = CafeSimulation(config_file)
        sim.run()

        metrics = sim.get_last_metrics()
        metrics["scenario"] = scenario_name
        scenario_results.append(metrics)

    except FileNotFoundError:
        print(f"  ‚ö†Ô∏è  Soubor {config_file} neexistuje - p≈ôeskakuji")

if scenario_results:
    df_scenarios = pd.DataFrame(scenario_results)

    print("\n" + "=" * 60)
    print("POROVN√ÅN√ç SC√âN√Å≈ò≈Æ:")
    print("=" * 60)

    # Zobraz kl√≠ƒçov√© metriky
    comparison = df_scenarios[
        [
            "scenario",
            "total_customers",
            "total_served",
            "total_reneged",
            "success_rate",
            "avg_barista_queue",
            "avg_barista_utilization",
            "avg_cashier_queue",
            "avg_cashier_utilization",
            "avg_seats_occupied",
            "avg_seats_utilization",
            "max_barista_queue",
            "max_cashier_queue",
            "max_seats_occupied"

        ]
    ]
    print(comparison.to_string(index=False))

    # Export do CSV
    # df_scenarios.to_csv("scenario_comparison.csv", index=False)
    # print("\nüíæ V√Ωsledky ulo≈æeny do: scenario_comparison.csv")
```
